### Задать вручную число карточек, имя входного файла

```python
# вручную задать параметр: число карточек в документе - СЧИТАЕТСЯ С ПОМОЩЬЮ ПОИСКА по тексту названия села и года
num = 70
# в кавычки скопировать название файла БЕЗ РАСШИРЕНИЯ
filename="666"
#
# эту строку не менять:
#
filename1 = "filename.txt"#filename.txt
```

```python
file = open(file=filename1, mode="r", encoding="utf-8")
indexes = [i for i in range(num)]
tablitsa = pd.DataFrame(index=[indexes], columns=["программа", "вопрос", \
                                                  "село", "год", \
                                                  "соб.1", "соб.2", \
                                                  "соб.3", "соб.4", \
                                                  "текст", "инф.1", "инф.2", \
                                                  "инф.3", "инф.4", ""])
informanty = set()
sobirately = set()
for i in indexes:
    line1 = file.readline().replace("\n","")
    line1 = line1.split("-")
    tablitsa.iloc[i, 0]=line1[0]
    tablitsa.iloc[i, 1]=line1[1]
    del line1
    line1 = file.readline().replace("\n","")
    if "'" in line1:
        line1 = line1.split("'") #обрабатываем 2 типа апострофов
    elif "’" in line1:
        line1 = line1.split("’")
    tablitsa.iloc[i, 2]=line1[0]
    tablitsa.iloc[i, 3]=line1[1]
    del line1
    line1 = file.readline().replace("\n","")
    line1 = line1.split(", ")
    for n in range(len(line1)):
        if not line1[n] in sobirately:
            sobirately.add(line1[n])
        col = 4 + n
        tablitsa.iloc[i, col]=line1[n]
    del col
    del line1
    newline = ""
    line = ""
    while newline != "\n":
        newline = file.readline()
        newline2 = newline.replace("\n","")
        line = line + newline2 + "@"
        if line[-2:] == "@@":
            break
    line = line[:len(line)-2]
    line2 = line[line.rfind("@") + 1:]
    line2 = line2.split(", ") #extracting informants
    line = line[:line.rfind("@")] # selecting text deleted +1
    for n in range(len(line2)):
        if not line2[n] in informanty:
            informanty.add(line2[n])
        col = 9 + n
        tablitsa.iloc[i, col]=line2[n]#appending informants
    tablitsa.iloc[i,8] = line #appending text to the column
    
    #DataFrame.reorder_levels(self, order, axis=0)
```
